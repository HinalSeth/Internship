# Q.2
In the above question, now scrape the following details of each product listed in first 3 pages 
of your search results and save it in a data frame and csv. In case if any product vertical has 
less than 3 pages in search results then scrape all the products available under that product 
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of 
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" 
and “Product URL”. In case, if any of the details are missing for any of the product then 
replace it by “-“.

```python
#Creating 10 Empty lists to store the fetched data.
Brand_Name=[] 
Name_of_the_Product=[] 
Rating=[]
No_of_Ratings=[] 
Price=[]
ReturnExchange=[] 
ExpectedDelivery=[] 
Availability=[] 
Details =[]
UR_3=[]

for k in range(3):
    #collecting urls of all the product's URL
    urls_3=driver.find_elements(By.XPATH,"//a[@class='a-size-base a-link-normal a-text-normal']")
    for i in urls_3[:]:
        UR_3.append(i.get_attribute('href'))

    nxt_button = driver.find_element(By.XPATH,'//li[@class="a-last"]')    #clicking to next page link                                     
    nxt_button.click()                                                   
    time.sleep(5)     
```

```python
for i in UR_3:
    driver.get(i)   #fetching data from Each Product's URL
    time.sleep(5)
    
    try:    #collecting brand name
        brand=driver.find_element(By.XPATH,"//table[@id='technicalSpecifications_section_1']/tbody/tr[1]/td")
        Brand_Name.append(brand.text)
    except:
        Brand_Name.append('-')
        
    try:    #collecting Product Name
        name=driver.find_element(By.XPATH,"//span[@id='productTitle']")
        Name_of_the_Product.append(name.text)
    except:
        Name_of_the_Product.append('-')

        
    try:    #collecting ratings
        rate=driver.find_element(By.XPATH,"//span[@class='a-size-base a-nowrap']/span")
        Rating.append(rate.text)

    except:
        Rating.append('-')
        
    try:     #collecting Total number of ratings
        num=driver.find_element(By.XPATH,"//div[@class='a-row a-spacing-medium averageStarRatingNumerical']/span")
        No_of_Ratings.append(num.text)
    except:
        No_of_Ratings.append('-')
        
    try:    #collecting Price
        prc=driver.find_element(By.XPATH,"//td[@class='a-span12']")
        Price.append(prc.text)
        
    except:
        Price.append('-')
        
    try:     #collecting Return/Exchange Policy Details
        ret=driver.find_element(By.XPATH,"//div[@id='RETURNS_POLICY']/span/div[2]/a")
        ReturnExchange.append(ret.text)
    except:
        ReturnExchange.append('-')
        
    try:    #Collecting Expected Delivery Date
        dlv=driver.find_element(By.XPATH,"//div[@id='dynamicDeliveryMessage']//div//b")
        ExpectedDelivery.append(dlv.text)
    except:    
        ExpectedDelivery.append('-')
        
    try:    #collecting Availability of the product
        avl=driver.find_element(By.XPATH,"//span[@class='a-size-medium a-color-success']")
        Availability.append(avl.text)
    except:    
        Availability.append('-')
        
    try:     #collecting Product Details
        det=driver.find_element(By.XPATH,"//div[@id='feature-bullets']/ul/li[4]")
        Details.append(det.text)
    except:    
        Details.append('-')
```

```python
print(len(Brand_Name),len(Name_of_the_Product),len(Rating),len(No_of_Ratings),len(Price),len(ReturnExchange),len(ExpectedDelivery),len(Availability),len(Details),len(UR_3))
```

```python
Product = pd.DataFrame({})   #creating a datafrmae
Product['Brand']=Brand_Name
Product['Name Of The Product']=Name_of_the_Product
Product['Rating']=Rating
Product['Number Of Ratings']=No_of_Ratings
Product['Price']=Price
Product['Return/Exchange']=ReturnExchange
Product['Expected Delivery']=ExpectedDelivery
Product['Availability']=Availability
Product['Product Details']=Details
Product['Product URL']=UR_3
Product
```
