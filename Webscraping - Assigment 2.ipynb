# Q.1

```python
driver.get("https://www.naukri.com/")     #getting the webpage
```

```python
#finding elements for job search bar
search_job = driver.find_element(By.ID,"qsb-keyword-sugg")     # Enter “Data Analyst” in “Skill, Designations, Companies” field
search_job.send_keys("Data Analyst")
search_loc = driver.find_element(By.XPATH,"//input[@id='qsb-location-sugg']") # enter “Bangalore” in “enter the location”
search_loc.send_keys("Bangalore")
search_btn = driver.find_element(By.XPATH,"//div[@class='search-btn']/button").click() #click the search button. 
```

```python
#we will create four empty lists now for job title, job location, company name and experience required
job_title=[]
job_loc=[]
comp_name=[]
ex_required=[]

#now we will get all the tags for title of job
title_tags = driver.find_elements(By.XPATH,"//a[@class='title fw500 ellipsis']")
title_tags[0:10]

#now we will get all the job titles in text format and append them in a list
for i in title_tags:
    title = i.text
    job_title.append(title)
job_title[0:10]
```

```python
#now we will get tags for location and append the text in list
loc_tags = driver.find_elements(By.XPATH,"//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
for i in loc_tags :
    loc=i.text
    job_loc.append(loc)
job_loc[0:10]   
```

```python
companies_tags=driver.find_elements(By.XPATH,"//a[@class='subTitle ellipsis fleft']") #getting data for company name
companies_tags[0:4]

for i in companies_tags:
    company_name=i.text
    comp_name.append(company_name)
comp_name[0:10]    
```

```python
#scraping data for experience required
ex_tags=driver.find_elements(By.XPATH,"//li[@class='fleft grey-text br2 placeHolderLi experience']/span[1]")
for i in ex_tags:
    exp=i.text
    ex_required.append(exp)
ex_required[0:10]  
```

```python
#we will create the data frame
jobs = pd.DataFrame({})
jobs['Title']=job_title[0:10]
jobs['Location']=job_loc[0:10]
jobs['Company']=comp_name[0:10]
jobs['Experience Required']=ex_required[0:10]
jobs
```

```python
driver = webdriver.Chrome(service=s)
driver.maximize_window()
driver.get("https://www.ambitionbox.com/")     #webpage
```

```python
slr_opt= driver.find_element(By.XPATH,"//a[@class='link salaries']").click()    #click on salary bar
```

```python
search_job = driver.find_element(By.XPATH,"//input[@class='tt-input']")         # enter data scientist in search field
search_job.send_keys("Data Scientist")
```

```python
enter_job = driver.find_element(By.XPATH,"/html/body/div/div/div/main/section/div[2]/div/span/div/div/div") 
enter_job.click()     #click on data scientist
```

```python
company_name=[]       #empty list for all the attributes
total_salary_record=[]
average_salary=[]
minimum_salary=[]
maximum_salary=[]
experience_required=[]

companytags = driver.find_elements(By.XPATH,"//div[@class='name']/a")    #company name
for i in companytags:
    c=i.text
    company_name.append(c)

tsr_tags=driver.find_elements(By.XPATH,"//div[@class='name']/span")      #total salary record
for i in tsr_tags:
    t=i.text
    t=t.split('on')[1]
    total_salary_record.append(t)
    
avg_sal_tags=driver.find_elements(By.XPATH,"//div[@class='average-indicator-wrapper']/p")  #average salary  
for i in avg_sal_tags:
    a=i.text
    average_salary.append(a)
    
min_sal_tags=driver.find_elements(By.XPATH,"//div[@class='salary-values']/div[1]")    #minimum salary
for i in min_sal_tags:
    min=i.text
    minimum_salary.append(min)    
    
max_sal_tags=driver.find_elements(By.XPATH,"//div[@class='salary-values']/div[2]")    #maximum salary
for i in max_sal_tags:
    max=i.text
    maximum_salary.append(max)
    
exp_req_tags=driver.find_elements(By.XPATH,"//div[@class='salaries sbold-list-header']")    #experience required
for i in exp_req_tags:
    e=i.text
    e=e.split('\n')[2]
    experience_required.append(e)    
    
DSJobs = pd.DataFrame({'Company Name':company_name,'Total Salary Record':total_salary_record,'Average Salary':average_salary,'Minimum Salary':minimum_salary,'Maximum Salary':maximum_salary,'Experience Required':experience_required})
DSJobs    
```

```python
driver.quit()     #close all the tab
```
