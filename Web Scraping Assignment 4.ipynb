# 1. 
Scrape the details of most viewed videos on YouTube from Wikipedia.


Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
import time  #importing libraries
import selenium
import pandas as pd 
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
s=Service("C:\\Users\\Hinal\\Downloads\\chromedriver_win32\\chromedriver.exe")  #opening chrome
driver = webdriver.Chrome(service=s)    
driver.maximize_window()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")     #getting the webpage

rank=[] #empty list
name=[] #empty list
artist=[] #empty list
upload =[] #empty list
views=[] #empty list

time.sleep(2)

#fetching rank
r=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[1]")
for i in r:
    rank.append(i.text)

#fetching name    
n=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[2]")
for i in n:
    name.append(i.text)
    
#fetching artist
a=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[3]")
for i in a:
    artist.append(i.text)

#fetching upload date    
u=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[5]")
for i in u:
    upload.append(i.text)    

#fetching views    
v=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[4]")
for i in v:
    views.append(i.text)    
    
MostViewed=pd.DataFrame({})    #creating dataframe
#Since there are two different tables , and our required data is in first, hence we fetch it using indexing 
MostViewed['Rank']=rank[0:30] 
MostViewed['Name of video']=name[0:30]
MostViewed['Artist']=artist[0:30]
MostViewed['Upload Date']=upload[0:30]
MostViewed['Views']=views[0:30]
MostViewed
```

# 2.
Scrape the details team India’s international fixtures from bcci.tv.

Url = https://www.bcci.tv/.

You need to find following details:
A) Match title (I.e. 1
st ODI)
B) Series
C) Place
D) Date
E) Time

```python
driver.get("https://www.bcci.tv/")  #getting the webpage
time.sleep(2)
#clicking on international
international=driver.find_element(By.XPATH,"/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]").click() 
time.sleep(2)
#clicking on fixtures 
fixture=driver.find_element(By.XPATH,"/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a").click()

match=[]  #empty list
series=[] #empty list
place=[]  #empty list
mdate=[]  #empty list
mtime=[]  #empty list
url=[]    #empty list

#fetching match title
m=driver.find_elements(By.XPATH,"//p[@class='fixture__additional-info']/strong")
for i in m:
    match.append(i.text)

#fetching series    
s=driver.find_elements(By.XPATH,"//span[@class='u-unskewed-text fixture__tournament-label u-truncated']")
for i in s:
    series.append(i.text)

#fetching Place    
p=driver.find_elements(By.XPATH,"//p[@class='fixture__additional-info']/span")
for i in p:
    place.append(i.text)

#fetching urls of each match to get date and time     
urls=driver.find_elements(By.XPATH,"//div[@class='js-list']/a")
for i in urls[:]:
    url.append(i.get_attribute('href')) 
        
for i in url:      #fetching data from Each Match's URL
    driver.get(i)   
    time.sleep(2)
    
    d=driver.find_element(By.XPATH,"//div[@class='mc-header-scorebox__datetime']/strong")  #fetching date
    mdate.append(d.text)
    
    t=driver.find_element(By.XPATH,"//div[@class='mc-header-scorebox__datetime']/span[1]")  #fetching time
    mtime.append(t.text)

Fixtures=pd.DataFrame({})    #creating Dataframe
Fixtures['Match Title']=match
Fixtures['Series']=series
Fixtures['Place']=place
Fixtures['Date']=mdate
Fixtures['Time']=mtime
Fixtures
```

# 5.
Scrape the details of trending repositories on Github.com.

Url = https://github.com/
    
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used


```python
driver.get("https://github.com/")    #getting the webpage
explore=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary").click()
time.sleep(2)
trend=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a").click()

ttl=[]     #empty list
rdesc=[]   #empty list
cc=[]      #empty list
lang=[]    #empty list

t=driver.find_elements(By.XPATH,"//h1[@class='h3 lh-condensed']/a") #fetching repository title
for i in t:
    ttl.append(i.text)

try:    
    r=driver.find_elements(By.XPATH,"//p[@class='col-9 color-fg-muted my-1 pr-4']")    #fetching description
    for i in r:
        rdesc.append(i.text)
except:
    rdesc.append('-')   #if description is not available, appending '-'
    
l=driver.find_elements(By.XPATH,"//div[@class='f6 color-fg-muted mt-2']/span[1]")  #fetching language used
for i in l:
    lang.append(i.text)
    
url=[] #empty list
urls=driver.find_elements(By.XPATH,"//h1[@class='h3 lh-condensed']/a") #fetching url of each repository
for i in urls[:]:
    url.append(i.get_attribute('href')) 
for i in url:  #from each urls fetching data
    driver.get(i)   
    c=driver.find_element(By.XPATH,"//div[@class='Layout-sidebar']/div/div[5]/div/h2/a/span") #fetching contributors count
    cc.append(c.text)    
    
Trending=pd.DataFrame({})   #creating the dataframe
Trending['Repository Title']=ttl
Trending['Repository Description']=rdesc
Trending['Contributors Count']=cc
Trending['Language Used']=lang
Trending
```
