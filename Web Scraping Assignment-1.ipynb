# 2. Write a python program to display IMDB’s Top rated 100 movies’ data  and make data frame. 
 (i.e. Name, IMDB rating, Year of release) 

```python
page=requests.get('https://www.imdb.com/chart/top')
page
soup=BeautifulSoup(page.content)

names=[]
for j in soup.find_all('td',class_='titleColumn'):
    j=j.text.split('\n')[2]
    j=j.strip('  ')
    names.append(j) 
names [0:100]
```

```python
year=[]
for j in soup.find_all('span',class_='secondaryInfo'):
    year.append(j.text)
year [0:100]
```

```python
rating=[]
for j in soup.find_all('td',class_='ratingColumn imdbRating'):
    rating.append(j.text.replace('\n',''))
rating[0:100]
```

```python
data_movies=pd.DataFrame({'Movie Names':names,'IMDB Ratings':rating,'Year of Release':year})
data_movies[0:100]
```

```python
c=[]
for i in soup.find_all('td',class_='rankings-block__banner--rating u-text-right'):
    i=i.text.split('\n')[1]
    i=i.strip(' ')
    c.append(i)
c    

d=[]
for i in soup.find_all('td',class_='table-body__cell u-text-right rating'):
    d.append(i.text)
d[0:9] 

rating_m1 = c + d[0:9]
print(str(rating_m1))
```

```python
loc=[]
for i in soup.find_all('div',class_='restnt-loc ellipsis'):
    loc.append(i.text)
loc
```

```python
rate=[]
for i in soup.find_all('div',class_='img-wrap'):
    rate.append(i.text)
rate
```

```python
imgurl_r=[]
for i in soup.find_all('img',class_='no-img'):
    imgurl_r.append(i['data-src'])
imgurl_r
```

```python
df_r=pd.DataFrame({'Restaurant Name':rn, 'Cuisine':c,'Location':loc,'Rating':rate,'Image URL':imgurl_r})
df_r
```

```python
wc=[]
for i in soup.find_all(['td']):
    wc.append(i.text)
wc[19::6]
```

```python
hd=[]
for i in soup.find_all('td',class_='hr'):
    hd.append(i.text)
hd 
```

```python
pr=[]
for i in soup.find_all('td',class_='prob'):
    pr.append(i.text)
pr
```
